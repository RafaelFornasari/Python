# 04. Tipos de dados

1.   Listas - São criadas quando [ ] (colchetes)
2.   Tuplas - São criadas quando ( ) (parênteses)
3.   Dicionários -> São criadas quando { } (chaves)

Estes tipos de dados podem receer qualquer tipo de informação, inclusive Operações Aritiméticas


````python
# Listas
Lista_Exemplo_01 = [1, 2, 3, 4]
Campo_Exemplo = 2


print('Impressão de todos os elementos da lista:',Lista_Exemplo_01)
print('Impressão somente de qual campo desejo:',Lista_Exemplo_01[2])
print('Impressão teste:',Lista_Exemplo_01[Campo_Exemplo])
````

````python
Lista_Exemplo_02 = ['Nome', 'Nomenome', 'Idade', 24, True]

print(Lista_Exemplo_02)
````

````python
Lista_Exemplo_03 = [1, 'Texto', 987, [3, 'Idade', 'Rafael']]

print(Lista_Exemplo_03)
````

````python
#Tuplas !!!! Após criadas, não podem ser alteradas

Tupla_Exemplo_1 = (1,2,4,5)

Tupla_Exemplo_1
````

````python
#Dicionários
          #Aqui mora o ouro das listas

Dicionario_01 = {
    'Index' : 'valor',
    'id_01' : 1,
    'id_02' : 3,
    'Nome' : 'Rafael',
    'Lista': Lista_Exemplo_01,
    'Tupla' : Tupla_Exemplo_1
}

Dicionario_01
````

# 05. Nomeação de variáveis

Podemos nomear variáveis e tipos de dados de diversas formas:

1. Declarar as variáveis em uma única linha de comando;
2. Atribuir um valor à diversas variáveis;
3. Declarar usando uma Lista;
4. Combinar variáveis;
5. Operar com operadores matemáticos;
6. Muitos outros;


````python
#01. Nomeando
Laranja, Melão, Limão = 1, 2, 3

print(Laranja, Melão, Limão)
````

````python
#02. Nomeando
Morango = Manga = Feijao = 100

print(Morango, Manga, Feijao, 100)
````

````python
# 03. Noemar com listas
Listas_carros_01 = ['VW','Audi','Mercedes','BMW']

Marca_01, Marca_02, Marca_03, Marca_04 = Listas_carros_01

print(Marca_01, Marca_02, Marca_03,Marca_04)
print(Marca_02)
````

````python
#04. Combinando variáveis

Nome = 'Rafael'
Sobrenome = 'Fornasari'

nome_completo = Nome + " " + Sobrenome

nome_completo
````

````python
#05. Observações

Investimento = 10000
taxa = '0.2'
taxa_ajustada = float(taxa)

juros = Investimento * taxa_ajustada

juros
````

#06. Tipo de Informação

Na programação, o tipo de dado é um conceito muito importante.
Variáveis podem armazenar dados de diferentes tipos, e diferentes tipos podem fazer coisas diferentes.

O Python tem os seguintes tipos de dados integrados por padrão, nestas categorias:

1.   Text Type: *str*
2.   Numeric Types: *int, float, complex*
3.   Sequence Types: *list, tuple, range*
4.   Mapping Type: *dict*
5.   Set Types: *set, frozenset*
6.   Boolean Type: *bool*
7.   Binary Types: *bytes, bytearray, momeryview*


````python
String = str('Ola Mundo')
Inteiro = int(10)
Float = float(100.05)
Complex = complex ( 1j )
Lista = list(('Maça', 'Morango'))
Tupla = tuple( ('A', 'B' ))
Range = range(6)
Dicionario = dict (Nome = 'Odemir', age = 29 )
Set = set( ('A', 'B', 'C'))
Frozet = frozenset( ( 'A', 'B', 'C' ) )
Boleano = bool( 5 )
Bytes = bytes( 5 )
ByteArray = bytearray( 5 )
MemoryView = memoryview( bytes(5) )


from datetime import datetime

Data = datetime.today().date()
````
