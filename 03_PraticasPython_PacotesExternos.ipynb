O datetime módulo fornece classes para manipulação de datas e horas

```python
type("01/01/2024")

type(01/01/2024)
```

```python
import datetime
```

```python
Dia_Hoje = datetime.datetime.today()

print(Dia_Hoje)
```

```python
print(datetime.datetime.today().date())

print(datetime.datetime.today().time())
```

```python
#Vou criar uma variavel com a data de hoje, atualizada

Data_hoje = datetime.datetime.today().date()
```

```python
# Extraindo 'pedaços' da data


Ano = Data_hoje.year

Mes = Data_hoje.month

Dia = Data_hoje.day

print(Ano, Mes, Dia)
```

```python
Data_Antiga = datetime.date(2022, 1, 1)

type(Data_Antiga)
```

```python
#Calculo entre datas

Data_Antiga - Data_hoje
```

```python
# Ajuste data para o FORMATO BRASILEIRO

Data_hoje.strftime('%d/%m/%y')
```

# Pacote Time - Funções para manipulação do tempo

```python
# Primeiro contato

import time
```

```python
# Usando SLEEP() - Função para fazer o código aguardar.
# Importar bibli 'time', usa função sleep()

print('Comecei agora...')
time.sleep(3)

print ('Terminei')
```

```python
Agora = time.localtime()
print(Agora)
print(type(Agora))
```

```python
time.strftime('%d/%m/%y, %H:%M:%S', Agora)
```

```python
# Situação Real

Time_texto = '21 June, 2011'

#Como vamos tratar?

time.strptime( Time_texto, '%d %B, %Y')

#  % B -> Para fazer a conversão..
```

# Math - Operações Matemáticas

```python
import math
```

```python
Tupla_Exemplo_01 = (10, 5, 20, 30,40)

print('- Rasreio do Mínimo : min(Tupla_Exemplo_01 =', min(Tupla_Exemplo_01))

print('- Rasreio do Máximo : max(Tupla_Exemplo_01 =', max(Tupla_Exemplo_01))
```

```python
print('- Exponenciação : pow(2,2) =', pow(2,2),'\n\n------------')

print('\n','- Potencialção : math.sqrt(81) =', math.sqrt(49),'\n\n------------')

print('\n','- Arredondamento para cima:','\n','Ceil() -> math.ceil()','\n','math.ceil(1,4) = ', math.ceil(1.4),'\n\n------------')

print('\n','- Arredondamento para baixo:','\n',
      'Floor() -> math.floor()','\n',
      'math.floor(1,4) = ', math.floor(1.4),
      '\n\n------------')
```
