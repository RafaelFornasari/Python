# Manipulando Listas

Listas são um, dos 4 tipos, de armazenadores de coleção de dados.  Os outros tipos são:
- Tuple
- Set
- Dictionary

Comandos mais utilizados para listas:

1. append() - adiciona um item ao final da lista
2. len() - Calcula o tamanho da lista
3. [] - acessa uma posição da lista
4. del() - Excluir um elemento
5. clear() - Limpar a lista
6. insert() - Para inserir um item de lista em um índice espeficicado 'lista.insert(posição_desejada, valor_a_ser_inserido)
7. extend() - Anexar elementos de outra lista à lista atual
8. remove() - Remove o item especificado - especificamente o item
9. pop() - Remove o indice especificado
10. sort() - Ordenar os valores
11. copy() - Faz uma cópia da lista
12. index() - Retorna o index do elemento da lista

```python
print(Lista_Vazia)
```

```python
#Insert

Lista_Vazia.insert(2,"Inserção")
print(Lista_Vazia)
```

```python
# Extend

Lista_01 = [0, 1, 2, 3, 4]
Lista_02 = [23, 25, 45, 67]

Lista_01.extend(Lista_02)
print(Lista_01)
```

```python
# Remove

Lista_01.remove(2)

print(Lista_01)
```

```python
# Pop - Remove pela POPsição
print(Lista_01)

Lista_01.pop(0) #Remove a primeira posição

print("Nova Lista agora", Lista_01)
```

```python
# Ordenação de Lista

Lista_ordencao = ["Rafael", "Alexandre", "Pedro", "Guilherme", "Matheus", "Gustavo"]

Lista_ordencao.sort()

Lista_ordencao

Lista_ordenacao_Numero = [0, 12, 44, 53, 2, 1, 45]

Lista_ordenacao_Numero.sort()
Lista_ordenacao_Numero


# Para trazer de traz pra frente usa o 'reverse = True' dentro do sort'. True com T maiúsculo

Lista_ordenacao_Numero.sort( reverse = True )
Lista_ordenacao_Numero
```

```python
# Copiar uma lista?


Lista_ordencao_nova = Lista_ordencao.copy()
Lista_ordencao_nova
```
