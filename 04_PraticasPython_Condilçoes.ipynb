# Condição IF

IF padrãozinho

```python
idade = 17

if idade >= 18:
  print('Maior de idade')

elif idade >= 15 and idade <= 17:
  print('Seu pai precisa autporiozar')

else:
  print('Menor de idade')
```

#FOR - Loops

For é usado para uma 'travessia sequencial'.
Ele corre uma lista, tupla, string, o que quer que for

```python
for Qualquercoisa in range(10):
  print(Qualquercoisa)
```

```python
for Qualquercoisa in range(10):
  print(Qualquercoisa * 2)
```

```python
#Começar do 1 eir até o 10

for Qualquercoisa in range(1, 10):
  print(Qualquercoisa)
```

```python
#Começar a partir do 5

for Qualquercoisa in range(5, 10):
  print(Qualquercoisa)
```

```python
# Do 1 ao 100, porém pulando de 5 em 5

for UmAoCem in range(1,100,5):
  print(UmAoCem)
```

## Outros usos do FOR

```python
Lista_for_00 = ['Brasil','Uruguai','Chile','Colombia','Argentina','Africa do Sul','Italia','Alemanha','França','Inglaterra']

print(
    'O que vamos fazer aqui? Vamos praticar um pouco de loop com essa lista pelo que parece \n\n'
)
print(
    Lista_for_00
)

print(
    '\nAcima temos a lista como str. Abaixo temos a lista como elementos\n'
)

for index in Lista_for_00:
  print(index)
```

```python
print('Podemos aplicar as manipulações de string nesse for de cima\n')

print(
    'Por exemplo: Deixar todos maiúsculos.\n'
)
for index in Lista_for_00:
  print(index.upper())

print('\n\n Deixar todos minúsculos\n')

for index in Lista_for_00:
  print(index.lower())

print('\n\n Buscaar as 3 primeiras letras \n')

for index in Lista_for_00:
  print(index.lower()[0:3])
```

```python
#

for Pais in Lista_for_00:
  if Pais == 'Brasil' or Pais == 'Uruguai' :
    print(Pais,'Aqui é top\n')

  elif Pais == 'Inglaterra':
    print(Pais,'Amount of cheese\n')

  else:
    print(Pais,'Jamais serão\n')
```
